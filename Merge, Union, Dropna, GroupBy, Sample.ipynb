```python
import pandas as pd
```

```python
# I caso Dos dataframe jugadores con sus posiciones vs jugadores y pais, coinciden el nombre de la columna
df1 = pd.DataFrame({
   'Jugadores': ["Federer", "Ronaldo", "Phelps", "Valenzuela"],
   'Posicion': ["Delantero", "Delantero", "Medio campista", "Defensa"]
})
 
df2 = pd.DataFrame({
   'Jugadores': ["Federer", "Ronaldo", "Phelps", "Sotelo"],
   'Pais': ['Peru', 'Brasil', 'USA', 'Venezuela']
})
```

```python
#inner
df=pd.merge(df1,df2,how="inner",on="Jugadores", indicator=True)
df
```

```python
#left
df=pd.merge(df1,df2,how="left",on="Jugadores", indicator=True)
df
```

```python
#right
df=pd.merge(df1,df2,how="right",on="Jugadores", indicator=True)
df
```

```python
#outer
df=pd.merge(df1,df2,how="outer",on="Jugadores", indicator=True)
df
```

```python
# II caso Dos dataframe jugadores con sus posiciones vs jugadores y pais, no coinciden el nombre de la columna
df1 = pd.DataFrame({
   'Jugadores': ["Federer", "Ronaldo", "Phelps", "Valenzuela"],
   'Posicion': ["Delantero", "Delantero", "Medio campista", "Defensa"]
})
 
df2 = pd.DataFrame({
   'Futbolista': ["Federer", "Ronaldo", "Phelps", "Sotelo"],
   'Pais': ['Peru', 'Brasil', 'USA', 'Venezuela']
})
```

```python
#outer Mostrar los dos dataframe
df=pd.merge(df1,df2,how="outer",left_on= 'Jugadores', right_on='Futbolista', indicator=True)
df
```

```python
############################ Union All
df1 = pd.DataFrame({
   'Jugadores': ["Trauco", "Advincula", "Yotun", "Lapadula"],
   'Posicion': ["Delantero", "Delantero", "Medio campista", "Defensa"],
    'Pais': ["Perú", "Perú", "Perú", "Perú"]
})

df2 = pd.DataFrame({
   'Jugadores': ["Alejandro", "Ronaldo", "Bebeto", "Lapadula"],
   'Posicion': ["Delantero", "Delantero", "Medio campista", "Defensa"],
    'Pais': ["Brasil", "Brasil", "Brasil", "Perú"]
})
```

```python
df1_union_df2=pd.concat([df1,df2])
df1_union_df2
```

```python
# Union All reindex
df1_union_df2=pd.concat([df1,df2],ignore_index=True)
df1_union_df2
```

```python
# union
df1_union_df2=pd.concat([df1,df2],ignore_index=True).drop_duplicates()
df1_union_df2
```

```python
#Groupby
df = pd.DataFrame({
    'category': ['A'] * 3 + ['B'] * 2,
    'name': ['A1 A1.1 A1.2', '23', 'A3', 'B1', 'B2'],
    'num': range(1, 6),
    'par': ['x1', 'x1', 'x1', 'x1', 'x1']
})

df
```

```python
df.groupby('category').count()
```

```python
# Agrupar en una fila los valores
df.groupby('category').agg({
    'name': lambda x: ', '.join(x),
    'num': lambda x: x.max(),
    'par': lambda x: ', '.join(x).split(',')[0]
})
```
